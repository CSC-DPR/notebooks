# Read And write EOPF CPM object

## Loading Module

```python
from eopf.product.store import NetCDFReader, ZarrWriter, ZarrReader
import zarr
import os
```

## Initialize parameters

```python
patterns_group = [(".*_coordinates.nc", "coordinates"), (".*_radiance.nc", "radiance")]
path = "eopf_product_data_structure/data/S3A_OL_1_EFR____20200101T101517_20200101T101817_20200102T141102_0179_053_179_2520_LN1_O_NT_002.SEN3/"
ZARR_DIR = "zarr_writting"
store = zarr.DirectoryStore(ZARR_DIR)
```

## Read Netcdf

```python
reader = NetCDFReader(path, groups_patterns=patterns_group)
product = reader.read()
product
```

## Write Zarr

```python
os.removedirs(ZARR_DIR)
writer = ZarrWriter(store)
writer.write(product)
os.listdir(ZARR_DIR)
```
