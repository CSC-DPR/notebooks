```python
"""
products : list of paths to S2 products (L1 and L2)
product_type : product type to be displayed
"""
def s2_display_product(products,product_type):

    for product in products:
        product_name = os.path.basename(product)
        if product_type in product_name:
                
            display(HTML("<h1>Product: "+product_name[:10]+"</h1>"))
            
            prod_files = glob.glob(product  + "/GRANULE/*/IMG_DATA/*.jp2") + \
                         glob.glob(product  + "/GRANULE/*/IMG_DATA/R10m/*.jp2") + \
                         glob.glob(product  + "/GRANULE/*/IMG_DATA/R20m/*.jp2") + \
                         glob.glob(product  + "/GRANULE/*/IMG_DATA/R60m/*.jp2")

            for file in prod_files:
                file_name = os.path.basename(file)
                
                if '.jp2' in file_name:
                    ds = xr.ds = xr.open_rasterio(file)
                
                    display(HTML("<h3>Group: "+file_name+"</h3>"))
                    display(ds)

            
```
